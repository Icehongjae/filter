```python
def to_distance(z):
    h0 = 70
    v = z * 3e10 # c=  3e10cm/s
    distance = (h0 * 1e3 * 1e2 / v)**-1
    return distance
```

```python
def to_arcsec_aperture(z = None, aperture = None, distance = None): # aperture : pc 
    dist = distance
    return aperture/dist*206265
```
